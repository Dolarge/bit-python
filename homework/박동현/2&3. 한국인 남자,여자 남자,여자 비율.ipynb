```python
import pandas as pd
```

```python
pop = pd.read_excel('./data/01. population_in_Seoul.xls', encoding='utf-8')
pop.head()
```

```python
pop_kor = pd.read_excel('./data/01. population_in_Seoul.xls',
                         header=2, 
                         usecols='B,D,H,I', 
                         encoding='utf-8')
pop_kor.head()
```

```python
pop_kor_columns = pop_kor.columns

pop_kor.rename(columns={
    pop_kor_columns[0] : '구별',
    pop_kor_columns[1] : '총인구',
    pop_kor_columns[2] : '한국인남자',
    pop_kor_columns[3] : '한국인여자',
}, inplace=True)

pop_kor.head()
```

```python
pop_kor.drop([0], inplace=True)
pop_kor.head()
```

```python
pop_kor.sort_values(by='한국인남자',ascending=False).head()
```

```python
pop_kor.sort_values(by='한국인여자',ascending=False).head()
```

```python
pop_kor['한국남자비율']=pop_kor['한국인남자']/pop_kor['총인구']*100
pop_kor['한국여자비율']=pop_kor['한국인여자']/pop_kor['총인구']*100
pop_kor.head()
```

```python
pop_kor.sort_values(by='한국남자비율',ascending=False).head()
```

```python
pop_kor.sort_values(by='한국여자비율',ascending=False).head()
```
