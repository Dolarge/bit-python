```python
import pandas as pd
```

```python
pop = pd.read_excel('./data/01. population_in_Seoul.xls', encoding='utf-8')
pop.head()
```

```python
pop_foreign = pd.read_excel('./data/01. population_in_Seoul.xls',
                         header=2, 
                         usecols='B,D,K,L', 
                         encoding='utf-8')
pop_foreign.head()
```

```python
pop_foreign_columns = pop_foreign.columns

pop_foreign.rename(columns={
    pop_foreign_columns[0] : '구별',
    pop_foreign_columns[1] : '총인구',
    pop_foreign_columns[2] : '외국인남자',
    pop_foreign_columns[3] : '외국인여자',
}, inplace=True)

pop_foreign.head()
```

```python
pop_foreign.drop([0], inplace=True)
pop_foreign.head()
```

```python
pop_foreign['외국인남자비율']=pop_foreign['외국인남자']/pop_foreign['총인구']*100
pop_foreign['외국인여자비율']=pop_foreign['외국인여자']/pop_foreign['총인구']*100
pop_foreign.head()
```

```python
pop_foreign.sort_values(by='외국인남자비율', ascending=False).head()
```

```python
pop_foreign.sort_values(by='외국인여자비율', ascending=False).head()
```
