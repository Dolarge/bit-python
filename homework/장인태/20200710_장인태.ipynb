```python
fp1 = np.polyfit(data_result['한국인여자수'], data_result['소계'], 1)   #인구수 x, 소계 y에 놓는 1차함수
fp1
```

```python
f1 = np.poly1d(fp1) # fp1의 정보를 이용해서( 기울기, 절편 ) 1차 함수 수식을 만들어 낸다.

# 함수에 들어갈 값을 만든다.
fx = np.linspace(100000, 700000, 100) # 10만부터 70만까지 100의 간격으로 정수 배열을 만들어 낸다.

plt.figure(figsize=(10, 10))
plt.scatter(data_result['한국인여자수'], data_result['소계'], s=50)

# 상관 계수를 활용한 직선 긋기
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('한국인여자수')
plt.ylabel('CCTV 소계')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['한국인여자수'], data_result['소계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

#오차 구하기   절댓값 씌움
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['한국인여자수']))

#오차별로 정령
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['한국인여자수'], data_result['소계'], c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')


for n in range(10):
    
    plt.text(df_sort['한국인여자수'][n]*1.02, df_sort['소계'][n] * 0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('한국인여자수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```

# 결론

 - 외국여자수는 상관관계가 관계없는 수준이여서 제외
 - 한국여자수는 상관관계가 0.3이상이라 약한 상관관계이지만 가설 실시
 - 인구수 대비 CCTV설치수와 비슷하게 결과 도출.
 - 인구수가 많은 자치구라면 그곳 역시 성별 상관없이 여성수가 많다
 - 따라서 여성수와 인구수는 비례하기때문에 여성수와 CCTV설치는 인구수와 비슷하다...?
