```python

pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls',
                          header = 2, # 엑셀상의 2번째 행 부터 데이터를 읽어 올 것-> 세 번째 행부터 데이터를 읽어온다.
                          usecols='G, H, I, J, K, L',
                          encoding='utf-8')
pop_Seoul.head()
```

```python
pop_seoul_columns = pop_Seoul.columns

pop_Seoul.rename(columns={
    pop_seoul_columns[0] : '한국인구수',
    pop_seoul_columns[1] : '한국인남자수',
    pop_seoul_columns[2] : '한국인여자수',
    pop_seoul_columns[3] : '외국인구수',
    pop_seoul_columns[4] : '외국인남자수',
    pop_seoul_columns[5] : '외국인여자수'
}, inplace=True)

pop_Seoul.head()
```

```python
pop_Seoul['외국인남자수비율'] = pop_Seoul['외국인남자수'] / pop_Seoul['한국인구수'] * 100
pop_Seoul['외국인여자수비율'] = pop_Seoul['외국인여자수'] / pop_Seoul['한국인구수'] * 100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='외국인남자수비율', ascending=False).head()
```
