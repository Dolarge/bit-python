```python
import pandas as pd
```

```python
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls', encoding='utf-8')
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls', 
                          header = 2, 
                          usecols = 'B,G,H,I',
                          encoding = 'utf-8' )

pop_seoul_columns = pop_Seoul.columns

pop_Seoul.rename(columns={ # 인구 데이터 컬럼명을 바꿔주었습니다
    pop_seoul_columns[1] : '한국인',
    pop_seoul_columns[2] : '남자',
    pop_seoul_columns[3] : '여자',
}, inplace=True)

pop_Seoul.sort_values(by='한국인', ascending=False).head() # 2번문제_한국인 남자/여자를 내림차순으로 정렬
```

```python
# 2번문제_한국인 남자/여자 '비율'을 내림차순으로 정렬하기
pop_Seoul['남자_비율'] = (pop_Seoul['한국인'] / pop_Seoul['남자']) * 100
pop_Seoul['여자_비율'] = (pop_Seoul['한국인'] / pop_Seoul['여자']) * 100
pop_Seoul.head()
```
