```python
import pandas as pd # 데이터를 위해 pandas를 먼저 불러오고,(as=>)이것을 pd로 부르겠습니다
```

```python
#파일의 원본은 이러합니다
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls', encoding='utf-8')
pop_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls', 
                          header = 2, #행 데이터를 '계'부터 나오게 설정했습니다
                          usecols = 'B,J,K,L',
                          encoding = 'utf-8' )
pop_Seoul.head()
```

```python
pop_seoul_columns = pop_Seoul.columns

pop_Seoul.rename(columns={ # 인구 데이터 컬럼명을 바꿔주었습니다
    pop_seoul_columns[1] : '등록외국인',
    pop_seoul_columns[2] : '남자',
    pop_seoul_columns[3] : '여자',
}, inplace=True)

pop_Seoul.head()
```

```python
# '외국인 남녀총합'을 기준으로 내림차순 정렬하였습니다.
pop_Seoul.sort_values(by='등록외국인', ascending=False).head()
```

```python
# 1번문제_외국인 남녀 '비율'을 내림차순으로 정렬하기
pop_Seoul['남자_비율'] = (pop_Seoul['등록외국인'] / pop_Seoul['남자']) * 100
pop_Seoul['여자_비율'] = (pop_Seoul['등록외국인'] / pop_Seoul['여자']) * 100
pop_Seoul.head()
```
