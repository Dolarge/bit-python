```python
Hw = pd.read_excel('./data/01. population_in_Seoul.xls',
                    header = 2,
                    usecols = 'B,D,H,I,K,L',
                    encoding = 'UTF-8')
```

```python
Hw_columns = Hw.columns

Hw.rename(columns = {
    Hw_columns[0] : '구별',
    Hw_columns[1] : '총인구수',
    Hw_columns[2] : '한국인남자',
    Hw_columns[3] : '한국인여자',
    Hw_columns[4] : '외국인남자',
    Hw_columns[5] : '외국인여자'
},inplace=True)

Hw.head()
```

```python
Hw['한국남자비율'] = Hw['한국인남자'] / Hw['총인구수'] * 100
```

```python
Hw['한국여자비율'] = Hw['한국인여자'] / Hw['총인구수'] * 100
```

```python
Hw['외국남자비율'] = Hw['외국인남자'] / Hw['총인구수'] * 100
```

```python
Hw['외국여자비율'] = Hw['외국인여자'] / Hw['총인구수'] * 100
```
