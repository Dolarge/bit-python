## 1. 외국인 남자/여자 비율을 내림차순으로 정렬

```python
pop_Seoul1 = pd.read_excel('./data/01. population_in_Seoul.xls',
                           header=2,
                           usecols='B,D,J,K,L',
                           encoding='utf-8')
pop_Seoul1.head()
```

```python
# 데이터 컬럼명 바꿔주기
pop_Seoul1_columns = pop_Seoul1.columns

pop_Seoul1.rename(columns={
    pop_Seoul1_columns[0]:'구별',
    pop_Seoul1_columns[1]:'총인구수',
    pop_Seoul1_columns[2]:'외국인',
    pop_Seoul1_columns[3]:'남자외국인',
    pop_Seoul1_columns[4]:'여자외국인'
}, inplace=True)

pop_Seoul1.head()
```

```python
pop_Seoul1['여자외국인비율'] = pop_Seoul1['여자외국인'] / pop_Seoul1['총인구수']*100
pop_Seoul1['남자외국인비율'] = pop_Seoul1['남자외국인'] / pop_Seoul1['총인구수']*100
```

```python
pop_Seoul1.head()
```

```python
pop_Seoul1.drop([0], inplace=True)
pop_Seoul1.head()
```

```python
pop_Seoul1['구별'].unique()
```

```python
len(pop_Seoul1['구별'].unique())
```

```python
pop_Seoul1.drop([26], inplace=True)
pop_Seoul1.tail()
```

```python
# 외국인 남자비율을 내림차순으로 정렬
pop_Seoul1.sort_values(by='남자외국인비율', ascending=False).head()
```

```python
# 외국인 여자비율을 내림차순으로 정렬
pop_Seoul1.sort_values(by='여자외국인비율', ascending=False).head()
```

## 3. 한국인 남자비율/여자비율을 내림차순으로 정렬

```python
pop_Seoul3 = pd.read_excel('./data/01. population_in_Seoul.xls',
                           header=2,
                           usecols='B,D,G,H,I',
                           encoding='utf-8')
pop_Seoul3.head()
```

```python
pop_Seoul3_columns=pop_Seoul3.columns

pop_Seoul3.rename(columns={
    pop_Seoul3_columns[0]:'구별',
    pop_Seoul3_columns[1]:'총인구수',
    pop_Seoul3_columns[2]:'한국인',
    pop_Seoul3_columns[3]:'남자한국인',
    pop_Seoul3_columns[4]:'여자한국인'
}, inplace=True)

pop_Seoul3.head()
```
